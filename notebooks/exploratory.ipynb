```python
import ast
import pandas as pd
from collections import Counter
import os
os.chdir("..")

df = pd.read_csv("data/processed/jobs_clean.csv")

def to_list(v):
    try:
        return ast.literal_eval(v) if isinstance(v, str) else v
    except Exception:
        return []

df["tags"] = df["tags"].map(to_list)

# 1. Top categorias
print(df["category"].value_counts().head(10))

# 2. Top tecnologias
all_tags = [tag for tags in df["tags"] for tag in tags]
print(Counter(all_tags).most_common(15))

# 3. Tipo de vaga
print(df["job_type"].value_counts())

# 4. % sem salário
sem_salario = df["salary"].isna().mean() * 100
print(f"{sem_salario:.1f}% das vagas não informam salário")
```
